```python
import numpy as np
import matplotlib.pyplot as plt

# 0부터 5까지 0.2씩 증가시킨 값 저장
t = np.arange(0, 5, 0.2)
t2 = t**2
t3 = t**3

# mu를 평균으로, sigma를 표준편차로 갖는 정규분포로부터 랜덤하게 값을 추출
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# 그래프 시각화
plt.figure(figsize=(10, 6))

# 1. 선 그래프
plt.subplot(2, 2, 1)
plt.plot(t, t, 'r--', label='Linear')
plt.plot(t, t2, 'bs', label='Quadratic')
plt.plot(t, t3, 'g^', label='Cubic')
plt.legend()
plt.title('Line Plot')

# 2. 히스토그램
plt.subplot(2, 2, 2)
plt.hist(x, bins=50, color='c', alpha=0.75)
plt.title('Histogram')

# 3. 산점도
plt.subplot(2, 2, 3)
plt.scatter(t, t2, color='m', label='t vs t^2')
plt.scatter(t, t3, color='y', label='t vs t^3')
plt.legend()
plt.title('Scatter Plot')

# 4. 막대 그래프 (예시)
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 4]
plt.subplot(2, 2, 4)
plt.bar(categories, values, color='orange')
plt.title('Bar Chart')

# 전체 그래프를 보여줌
plt.tight_layout()
plt.show()
```
